A(n)을 학생이 n명일 때 위 문제의 조건을 만족하는 경우의 수라고 했을 때, A(1)=1, A(2)=2를 만족한다. n명의 학생이 있을 때, A(n)은 다음의 두 가지 경우로 나눌 수 있다.
1. x라는 학생이 y라는 학생의 휴대폰을 가져갔을 때, y학생이 x학생의 휴대폰을 가져간 경우: 이 경우는 (n-2)명의 학생이 모두 자신의 휴대폰이 있는 집합에서 어느 누구도 본인의 휴대폰을 가져가지 않을 경우의 수이기 때문에 A(n-2)와 같다. x학생이 가져갈 수 있는 휴대폰의 가지 수는 본인의 것을 제외한 (n-1)가지이다. 따라서 이 경우 (n-1)*A(n-2)가지이다.

2. x라는 학생이 y라는 학생의 휴대폰을 가져갔을 때, y학생이 x학생의 휴대폰이 아닌 다른 학생의 휴대폰을 가져간 경우: 이 경우에는 x학생을 제외한 모든 학생이 가져갈 수 있는 휴대폰의 경우는 본인의 휴대폰을 제외한 (n-1)가지이다. 따라서 A(n-1)의 경우의 수 이고, x학생이 가져갈 수 있는 휴대폰의 가지 수는 본인의 것을 제외한 (n-1)가지이다. 따라서 이 경우 (n-1)*A(n-1)가지이다.

두 경우를 합한 것이 A(n)의 모든 경우의 수 이므로, A(n) = (n-1) * A(n-2) + (n-1) * A(n-1) = (n-1) * (A(n-2) + A(n-1)) 이다.

카페에서 커피를 배달할 때, 직원과 배달 로봇을 사용할 수 있다. 직원과 배달 로봇은 충분히 많다.
